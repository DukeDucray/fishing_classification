# Using Sample data in API

```python
import os
SERVICE_URL = os.environ.get("SERVICE_URL")
print(SERVICE_URL)
```

```python
import requests
import pandas as pd

# URL of your API endpoint
base_url = SERVICE_URL
endpoint = "/predict/
api_url = f"{base_url}{endpoint}

path = "../API/data/sample_data"
file_name = 'sample_purse_seines'

data = pd.read_csv(f'{path}/{file_name}.csv')

# Convert data to list of dictionaries
json_data = data.to_dict(orient='records')

# Make a POST request to the API
response = requests.post(api_url, json=json_data)

# Check the response
if response.status_code == 200:
    result = response.json()
    print(result)
else:
    print("Error:", response.text)
```

```python
subset = pd.DataFrame(result)
```

```python
subset
```

```python
import matplotlib.pyplot as plt

plt.plot(subset['lat'], subset['lon'], label='track')
plt.scatter(subset['lat'], subset['lon'], c=subset['is_fishing'], cmap='coolwarm', label='is_fishing')
plt.title(f'Around is_fishing occurrence')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

plt.legend()
plt.show()
```
